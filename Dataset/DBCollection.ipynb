```python
test_num=153
le=''
test=keypoint[test_num]
for j in file_name[test_num]:
    if j != 0:
        le=le+chr(j)
list_na=[]
for u in range(11):
    na=''
    for k in keypoint_name[u]:
        if k != 0:
            na=na+chr(k)
    list_na.append(na)
```
