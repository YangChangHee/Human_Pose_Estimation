```python
import dbcollection as dbc
import cv2
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm, tqdm_notebook
from tqdm import trange
import time
```

```python
LSPE=dbc.load('leeds_sports_pose_extended')
```

```python
LSPE.info()
```

```python
def Visualize_LSP(LSP,Kind,Num):
    keypoint=LSP.get(Kind,'keypoints')
    file_name=LSP.get(Kind,'image_filenames')
    le=''
    test=keypoint[Num]
    for i in file_name[Num]:
        le=le+chr(i)
    img=cv2.imread(le)
    test=np.array(test,dtype=np.uint8)
    print(test)
    for i in test:
        cv2.putText(img, '*',(i[0],i[1]),cv2.FONT_HERSHEY_SCRIPT_SIMPLEX,0.2,(0,255,0))
    plt.imshow(img)
    print(img.shape)
```

```python
Visualize_LSP(LSPE,'train',1520)
```

```python
def make_dataset_LSP(LSP,k):
    kp_data=LSP.get(k,'keypoints')
    keypoint_name=LSP.get(k,'keypoint_names')
    file_name=LSP.get(k,'image_filenames')
    
    ###### Keypoint_Num and Name ########
    kp_name=[]
    for u in tqdm(range(14)):
        time.sleep(0.0000001)
        na=''
        for k in keypoint_name[u]:
            if k!=0:
                na=na+chr(k)
        kp_name.append(na)
        
    ######### Keypoint ##########
    
    
    ######### Image ############
    
    img_data=[]
    for i in tqdm(file_name):
        time.sleep(0.0000000001)
        le=''
        for j in i:
            if j != 0:
                le=le+chr(j)
        img=cv2.imread(le)
        img_data.append(img)
    img_data=np.array(img_data)
    kp_data=np.array(kp_data)
    kp_name=np.array(kp_name)
    return img_data, kp_data, kp_name
```

```python
img_data, kp_data, kp_name =make_dataset_LSP(LSPE,'test')
```
