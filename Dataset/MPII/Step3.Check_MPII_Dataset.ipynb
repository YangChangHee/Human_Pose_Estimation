```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('mpii_dataset.csv')
```

```python
df["NAME"][0]
```

```python
def make_kp(df,index):
    kp_list=[]
    px=2
    kp_name=[]
    while px<34:
        col=df.columns[px]
        px=px+2
        kp_name.append(col[:-2])
    px=2
    while px<34:
        mk_kp=[]
        col=df.columns[px]
        x=df[col][index]
        col=df.columns[px+1]
        y=df[col][index]
        mk_kp.append(x)
        mk_kp.append(y)
        px=px+2
        kp_list.append(mk_kp)
    return np.array(kp_list), kp_name
```

```python
kp,name=make_kp(df,1)
```

```python
import cv2
def check_image(index):
    image_path='/home/lab/MPII_COCO/images/'
    img_name=df["NAME"][index]
    print("Image_Name : ",img_name)
    img=image_path+img_name
    img=cv2.imread(img)
    img=np.array(img)
    print("Image Size : ",img.shape)
    plt.imshow(img)
    
    kp,name=make_kp(df,index)
    print(kp)
    for i,k in kp:
        plt.scatter(i,k)
    for i,j in enumerate(name):
        print(i+1,"keypoint_name : ", j)
```
